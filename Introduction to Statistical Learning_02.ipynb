```r
#seq
x <- seq(1, 10)
x
x <- 1:10
x 
x <- seq(-pi, pi, length=50)
x
```

```r
y <- x
#outer is the outer product (matrix) of two arrays, with optional function transform
f <- outer(x, y, function(x, y) cos(y) / (1 + x^2)) 
#f
#contour(x, y, f)
#contour(x, y, f, nlevels=45, add=T)
fa <- (f-t(f))/2
contour(x, y, fa, nlevels=15)
```

```r
?outer
```

```r
image(x, y, fa) 
persp(x, y, fa) 
persp(x, y, fa, theta=30)
persp(x, y, fa, theta=30, phi=20)
persp(x, y, fa, theta=30, phi=70)
persp(x, y, fa, theta=30, phi=40)
```

```r
Auto <- read.csv("Auto.csv")
#View(Auto)
head(Auto)
Auto <- read.csv("Auto.csv", header=T, na.strings="?", stringsAsFactors=T)
head(Auto)
dim(Auto)
```

```r
plot(Auto$cylinders, Auto$mpg)
attach(Auto)
plot(cylinders, mpg)
```

```r
#change cylinders to qualitative variable
cylinders = as.factor(cylinders)
plot(cylinders, mpg)
plot(cylinders, mpg, col="red") 
plot(cylinders, mpg, col="red", varwidth=T) 
plot(cylinders, mpg, col="red", varwidth=T, horizontal = T)
plot(cylinders, mpg, col="red", varwidth=T, horizontal = T, xlab="cylinders", ylab="MPG")
```
